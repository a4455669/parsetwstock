```python
def CrawlInfoData(inputdate):
    url = "http://www.twse.com.tw/exchangeReport/BWIBBU_d?response=json&date=" + inputdate + "&selectType=ALL"
    with urllib.request.urlopen(url) as tempurl:
        tempdata = json.loads(tempurl.read().decode())
    
    try:
        if tempdata['date'] == inputdate:
            logger.info("<<CrawlInfoData-"+inputdate + '>>: ' + tempdata['stat'])
            return tempdata
    except KeyError:
        logger.error("<<CrawlInfoData-"+inputdate + '>>: ' + tempdata['stat'])
```

```python
def CrawlPriceData(inputmonth, inputNo):
    url = "http://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date=" + inputmonth + "&stockNo=" + inputNo
    with urllib.request.urlopen(url) as tempurl:
        tempdata = json.loads(tempurl.read().decode())
    
    try:
        if tempdata['date'] == inputmonth:
            logger.info("<<CrawlPriceData-"+inputmonth + '>>: ' + tempdata['stat'])
        return tempdata
    except KeyError:
        logger.error("<<CrawlPriceData-"+inputmonth + '>>: ' + tempdata['stat'])
    
```
